```python
import func


def WaveInfo_ReadCount(pid):
    func.TableBuild(pid)
    func.BinaryRead()
    func.Calculate()
    record = func.data.record_1
    return record

class Resp():
    def __init__(self, p_s, p_e):
        self.__p_s = p_s
        self.__p_e = p_e
        self.vent_len = p_e - p_s
        self.valid = True
        self.s_P = None
        self.s_F = None
        self.s_V = None
    
    def WaveInit(self, P, F, V):
        self.s_P = P[self.__p_s, self.__p_e]
        self.s_F = F[self.__p_s, self.__p_e]
        self.s_V = V[self.__p_v, self.__p_v]

flow_wh = WaveInfo_ReadCount(3782184)
plot_main = func.PlotMain()
```

```python
resp_range = slice(645, 670, 1)
s_P = func.np.array([])
s_F = func.np.array([])
s_V = func.np.array([])
for i in range(flow_wh.p_start[645], flow_wh.p_start[670]):
    s_P = func.np.append(s_P, flow_wh.s_P[i])
    s_F = func.np.append(s_F, flow_wh.s_F[i])
    s_V = func.np.append(s_V, flow_wh.s_V[i])

func.plt.figure(figsize=(16, 10))
func.plt.subplot(3, 1, 1)
func.plt.plot(s_P)
func.plt.subplot(3, 1, 2)
func.plt.plot(s_F)
func.plt.subplot(3, 1, 3)
func.plt.plot(s_V)
func.plt.show()
```
