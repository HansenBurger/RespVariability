```python
import func

def WaveInfo_ReadCount(pid):
    func.TableBuild(pid)
    func.BinaryRead()
    func.Calculate()
    outcome = func.Agger()
    return outcome.resp_t, outcome.rr

def WavePlot(x, y, df):
    func.sns.set_theme(style='whitegrid')
    func.sns.lmplot(x=x,
                   y=y,
                   data=df,
                   fit_reg=False,
                   height=3.3,
                   aspect=5.5)
    func.plt.show()
```

```python
ind, tar = WaveInfo_ReadCount(1157905)
slice_cut = slice(300, 550, 1)
len(ind)
```

```python
interp_rate = 100  # Hz
resample_rate = 4  # Hz

p_freq = func.FreqPreMethod(ind, tar, slice_cut)
p_freq.InitTimeSeries()
p_freq.InterpValue(interp_rate)
p_freq.Resampling(resample_rate)
```

```python
WavePlot('time', 'value', p_freq.df_sample)
```
